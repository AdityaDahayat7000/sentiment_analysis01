```python
df = df.drop(columns=['Unnamed: 0']) #Here we are dropping this column
```

```python
import re
import string
```

```python
# Function to clean the text of the tweets.

def clean_text(text) :
    text = re.sub(r'http\S+', '', text)  # Remove URLs
    text = re.sub(r'@\w+', '', text)  # Remove mentions
    text = re.sub(r'#\w+', '', text)  # Remove hashtags
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = text.translate(str.maketrans('', '', string.punctuation))  # Remove punctuation
    text = text.lower()  # Convert to lowercase
    return text

```

```python
df['cleaned_text'] = df['post_text'].apply(clean_text)
df.head()
```

```python
# Now we will perform some sentiment analysis on our data.
```

```python
pip install vaderSentiment
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
analyzer_1 = SentimentIntensityAnalyzer()
```

```python
#Function to get sentiment scores
```

```python
def get_sentiment_score(text):
    sentiment_score = analyzer_1.polarity_scores(text)
    return sentiment_score
```

```python
df['sentiment_scores'] = df['cleaned_text'].apply(get_sentiment_score)
df.head()
```

```python
#Creating columns for positive, neutral and negative comments

df['compound'] = df['sentiment_scores'].apply(lambda score_dict: score_dict['compound'])
df['positive'] = df['sentiment_scores'].apply(lambda score_dict: score_dict['pos'])
df['negative'] = df['sentiment_scores'].apply(lambda score_dict: score_dict['neg'])
df['neutral'] = df['sentiment_scores'].apply(lambda score_dict: score_dict['neu'])

```

```python
# Function to categorize sentiment based on compound score
```

```python
def categorize_sentiment(compound_score):
    if compound_score >= 0.05:
        return 'Positive'
    elif compound_score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'


df['sentiment_label'] = df['compound'].apply(categorize_sentiment)
df
```

```python
# Filter positive tweets
positive_tweets = df[df['sentiment_label'] == 'Positive']
print("Positive Tweets:")
print(positive_tweets[['post_text', 'sentiment_label']].head())  # Display the first few positive tweets

# Filter negative tweets
negative_tweets = df[df['sentiment_label'] == 'Negative']
print("Negative Tweets:")
print(negative_tweets[['post_text', 'sentiment_label']].head())  # Display the first few negative tweets
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='sentiment_label', data=df)
plt.title('Sentiment Analysis')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
 # Trend of positive comments with time

df['post_created'] = pd.to_datetime(df['post_created'], errors='coerce')


# Filter for positive tweets
positive_tweets = df[df['sentiment_label'] == 'Positive']

positive_tweets['date'] = positive_tweets['post_created'].dt.date


daily_positive_counts = positive_tweets.groupby('date').size()


plt.figure(figsize=(12, 6))
daily_positive_counts.plot(kind='line', marker='o', color='green')
plt.title('Trend of Positive Tweets Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Positive Tweets')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```
